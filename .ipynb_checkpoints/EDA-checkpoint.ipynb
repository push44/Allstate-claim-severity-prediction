## A) Data Information <br>
This data is downloaded from kaggle competition https://www.kaggle.com/c/allstate-claims-severity. This is an anonimous data as it is regarding the clients of Allstate insurance company. The train data set contains 188318 rows where each row represents single client. We have 130 number of predicators of which 16 are numerical and 114 are categorical. Our task here is to predict the severity of a claim of any given client based on the information provided by the 130 predictors.

## B) Background <br>
Once any client takes insurance from a company, client has to start paying premiums to the company according to defined terms and condition. Company collects many such premiums from its clients and invests that amount into profitable deals. When any of the comapany's clients faces undesirable expenses, the company pays amounts of those expenses according predefined conditions. Here, Allstate tries to predict the severity of the claim for more cost efficient functioning of the company. Hence, our target variable is a continous variable because it some how represents the claims severity. But it worth to note that units of these claims severity is unknown.

## C) ML Problem Formulation <br>
Since, we have continous values to predict this is a regression problem. Whether the entire data set shows linear trend with the target feature or not reamins to be seen but considering 130 predictor feature tree based algorithms seems to be a good choice here.

## D) Performance Metric: <br>
Competition considers Mean Absolute Error (MAE) as the performance metric. <br>
MAE=$\sum_{i}(y_i-\hat y_i)$

## E) Data Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')
import warnings
warnings.filterwarnings('ignore')
from scipy.stats import skew
import seaborn as sns
```

```python
# load the data set and set loss as new variable
train=pd.read_csv('inputs/train.csv')
print(train.head())
loss=train.loss
train.drop(['id','loss'],axis=1,inplace=True)
```

```python
# seperate train and test features
num_feat=train.select_dtypes(exclude='object')
cat_feat=train.select_dtypes(include='object')
```

```python
# check nan values in data
print(dict(train.isnull().sum())))
```

No column has null i.e., nan values. So we are good to go.

```python
# check loss variable
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(15,6))
ax[0].hist(loss,bins=500)
ax[0].set_ylabel('Frequency')
ax[0].set_xlabel('loss values')
ax[0].set_title('Histogram of loss variable')

ax[1].boxplot(loss)
ax[1].set_ylabel('loss values')
ax[1].set_title('Boxplot of loss variable')
print('Min value:',min(loss),'Max value:',max(loss));
```

As we can see, in the above histogram, the distribution looks very right skewed. Also, boxplot shows mean would be higher than a median value, shown by second quartile (blue) line by boxplot. Hence, we have to scale and shift the loss variable, in feature scaling step by using to appropriate transformation.

```python
# first downcast entire data set to save RAM. credit:https://www.kaggle.com/anshuls235/m5-forecasting-eda-fe-modelling
def downcast(df):
    cols = df.dtypes.index.tolist()
    types = df.dtypes.values.tolist()
    for i,t in enumerate(types):
        if 'int' in str(t):
            if df[cols[i]].min() > np.iinfo(np.int8).min and df[cols[i]].max() < np.iinfo(np.int8).max:
                df[cols[i]] = df[cols[i]].astype(np.int8)
            elif df[cols[i]].min() > np.iinfo(np.int16).min and df[cols[i]].max() < np.iinfo(np.int16).max:
                df[cols[i]] = df[cols[i]].astype(np.int16)
            elif df[cols[i]].min() > np.iinfo(np.int32).min and df[cols[i]].max() < np.iinfo(np.int32).max:
                df[cols[i]] = df[cols[i]].astype(np.int32)
            else:
                df[cols[i]] = df[cols[i]].astype(np.int64)
        elif 'float' in str(t):
            if df[cols[i]].min() > np.finfo(np.float16).min and df[cols[i]].max() < np.finfo(np.float16).max:
                df[cols[i]] = df[cols[i]].astype(np.float16)
            elif df[cols[i]].min() > np.finfo(np.float32).min and df[cols[i]].max() < np.finfo(np.float32).max:
                df[cols[i]] = df[cols[i]].astype(np.float32)
            else:
                df[cols[i]] = df[cols[i]].astype(np.float64)
        elif t == np.object:
            if cols[i] == 'date':
                df[cols[i]] = pd.to_datetime(df[cols[i]], format='%Y-%m-%d')
            else:
                df[cols[i]] = df[cols[i]].astype('category')
    return df

num_feat=downcast(num_feat)
cat_feat=downcast(cat_feat)
loss=downcast(pd.DataFrame(loss))
```

```python
# check dimmensions of features
num_feat.shape,cat_feat.shape
```

```python
# check the ditributions of numeric features
fig,ax=plt.subplots(nrows=7,ncols=2,figsize=(15,15))
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
num_cols=num_feat.columns
for ind in range(0,len(num_feat.columns),2):
    # plot two figures in same row
    num_feat[num_cols[ind]].hist(ax=ax[ind//2][0])
    num_feat[num_cols[ind+1]].hist(ax=ax[ind//2][1])
    # give titles both figures
    ax[ind//2][0].set_title(num_cols[ind])
    ax[ind//2][1].set_title(num_cols[ind+1])
fig.suptitle('Distribution plots of numeric features',fontsize=16);
```

We see distributions are not normal and as we know many models have assumtion of normality.

```python
# check the skewness of the features
skew_list=[]
for feat in num_cols:
    skew_list.append(skew(num_feat[feat]))
    
fig,ax=plt.subplots(figsize=(9,9))
ax.scatter(range(len(skew_list)),skew_list)
ax.plot(skew_list)
ax.set_xticks(range(len(num_cols)))
ax.set_xticklabels(num_cols,rotation=45)
ax.set_ylabel('skewness',fontsize=12)
ax.set_title('Feature skewness');
```

As expected features are skewed which is evident from the plot before this, but this plot shows the actual skewness values. Our ideal bet would be have skewness equal to zero as normal distribution is zero skewed. We will do this in the data preparation step where we will unskew features as much as possible for better model performance.

```python
# check linear correlation of loss variable with numeric features
fig,ax=plt.subplots(nrows=7,ncols=2,figsize=(15,15))
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
for ind in range(0,len(num_cols),2):
    # plot two figures in same row
    ax[ind//2][0].scatter(num_feat[num_cols[ind]],loss)
    ax[ind//2][1].scatter(num_feat[num_cols[ind+1]],loss)
    # give titles to both figures
    ax[ind//2][0].set_title(num_cols[ind])
    ax[ind//2][1].set_title(num_cols[ind+1])
fig.suptitle('Correlation plots of loss variable with numeric features',fontsize=16);
```

None of the numeric features is strongly-linearly correlated with the loss variable

```python
# find corelation coefficients between numeric features
corr=num_feat.corr()
fig,ax=plt.subplots(figsize=(12,12))
sns.heatmap(corr,cmap='YlGnBu',annot=True,cbar=False,ax=ax)
ax.set_title('Feature Correlation matrix',fontsize=16);
```

We see some features have correlation togather. We will not want to feed our models with correlated features, that will hamper model performance. In feature selection step we want to remove correlated features.

```python
# find highly correlated features
arr=np.array(corr[:7])
dep_feat=[]
for ind in range(arr.shape[0]):
    ar=pd.Series(arr[ind][ind+1:])
    for feat in num_cols[np.array(ar[ar>=0.7].index+(ind+1))]:
        dep_feat.append((num_cols[ind],feat))
        print('Correlation between features',num_cols[ind],feat,':',corr.iloc[ind][feat])
```

Above feature pairs show strong linear dependence with themselves

```python
# check linear dependece of features
fig,ax=plt.subplots(nrows=5,ncols=2,figsize=(15,15))
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
for ind in range(0,len(dep_feat),2):
    # plot two figures in same rows
    ax[ind//2][0].scatter(num_feat[dep_feat[ind][0]],num_feat[dep_feat[ind][1]])
    ax[ind//2][1].scatter(num_feat[dep_feat[ind+1][0]],num_feat[dep_feat[ind+1][1]])
    # give x and y axis labels to both figures
    ax[ind//2][0].set_xlabel(dep_feat[ind][0])
    ax[ind//2][1].set_ylabel(dep_feat[ind][1])
    
    ax[ind//2][0].set_xlabel(dep_feat[ind+1][0])
    ax[ind//2][1].set_ylabel(dep_feat[ind+1][1])
fig.suptitle('Pair plots of correlated features',fontsize=16);
```

```python
# check numeber of categories in categorical features
fig,ax=plt.subplots(figsize=(15,8))
feat_level=[]
cat_cols=cat_feat.columns
for feat in cat_cols:
    feat_level.append(len(cat_feat[feat].unique()))
ax.hist(feat_level)
ax.set_ylabel('Frequency',fontsize=16)
ax.set_xlabel('Number of categories',fontsize=16)
ax.set_title('Frequency plot of categories in the features',fontsize=16);
```
